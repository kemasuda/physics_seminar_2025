```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# プロットの調整
import seaborn as sns
sns.set(style='ticks', font_scale=1.6, font='times')
plt.rcParams["figure.figsize"] = (11,6)
```

```python
d = pd.read_csv("exoplanets.csv")
```

```python
d
```

```python
m, a, disc = np.array(d['mass']), np.array(d['semi-major axis']), np.array(d['discoverymethod'])
```

```python
idx_transit = (disc == 'Transit')
idx_doppler = (disc == 'Radial Velocity')
```

```python
dsol = pd.read_csv("solar_system.txt", comment='#')
```

```python
dsol
```

```python
plt.yscale("log")
plt.xscale("log")
plt.ylabel("planetary mass (Earth mass)")
plt.xlabel("semi-major axis (au)")
plt.plot(a[idx_doppler], m[idx_doppler], '.', label='Radial Velocity', mfc='none', mew=1)
plt.plot(a[idx_transit], m[idx_transit], 's', label='Transit', mfc='none', markersize=3, mew=0.5)
plt.plot(dsol["sma_au"], dsol["mass_earth"], 'D', label='Solar System', markersize=5, mec='k', mew=0.5)
plt.legend(loc='lower right');
```
